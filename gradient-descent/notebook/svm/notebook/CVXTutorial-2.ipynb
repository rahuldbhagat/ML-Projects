```python
A = np.array([[-2,-1], [-1,-2]])
A
```

```python
b=np.array([-8,-10]).T
b
```

```python
c=np.array([50,70]).T
c
```

```python
x = cp.Variable(2)
prob = cp.Problem(cp.Minimize(c.T@x),
                 [A @ x <= b])
prob.solve()
# Print result.
print("\nThe optimal value is", prob.value)
print("A solution x is")
print(x.value)
```
